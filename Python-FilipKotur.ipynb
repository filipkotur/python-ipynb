```python
import nltk
import string
import re
import string
from nltk.stem import PorterStemmer  
from nltk import sent_tokenize, word_tokenize
from nltk.stem.snowball import SnowballStemmer
from nltk.text import Text  
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.probability import FreqDist
from nltk.corpus import stopwords
from nltk.collocations import BigramCollocationFinder 
from nltk.metrics import BigramAssocMeasures 
from nltk.corpus import brown
from nltk.corpus import *
import string
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
from nltk.corpus import stopwords
import nltk
from nltk import corpus
from nltk import Text
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize
from nltk.corpus import movie_reviews
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import os
from nltk.tokenize import word_tokenize
from nltk import pos_tag, RegexpParser 

#Ucitanje teksta 
text = open("C:\\Users\\filip\\Desktop\\New folder (2)\\tekstreviews.txt", "r")

raw = text.read()

tokens = nltk.word_tokenize(raw)

#NORMALIZACIJA
#Sav tekst u mala slova
tokens = raw.lower()
#TOKENIZACIJA
#sve se pretvara u tokene
tokens = nltk.word_tokenize(tokens)
print(tokens)



```

```python

#brisanje dodatnih znakova
dodatni_znakovi = ['[', '}', '{',']','br','—','…','–','``',"...","''"]
punctuation = list(string.punctuation)

punctuation.extend(dodatni_znakovi)
lista = []
for x in tokens:
    if x not in punctuation:
        lista.append(x)

print(sorted(set(lista)))
```

```python
#stemizacija

steems = []
for a in lista:
    steems.append(nltk.PorterStemmer().stem(a))
print(steems)
```

```python
#Lematizacija

nltk_lemmaList = []
for x in steemed:
    nltk_lemmaList.append(nltk.WordNetLemmatizer().lemmatize(x))

print(nltk_lemmaList)
```

```python
#Uklanjanje STOP WORD
filtrirane_recenice = []  

nltk_stop_words = set(stopwords.words("english"))

for x in nltk_lemmaList:  
    if x not in nltk_stop_words:  
        filtrirane_recenice.append(x)  

print(filtrirane_recenice)
```

```python
#Izracunavanje frekvencija najcescih 30 rijeci i prikazvanje tekstualno i graficki


prva_analiza = nltk.FreqDist(filtrirane_recenice)
prva_analiza.most_common(25) 
```

```python
izbacujem = ["'s","wa"]
sve_rijeci = dict([(m, n) for m, n in prva_analiza.items() if m not in izbacujem]) 
nltk.FreqDist(sve_rijeci).plot(30, cumulative=False)
```

```python
#Prikazivanje Concordance za 3 najcesce rijeci u tekstu
 

Konkordancija = Text(filtrirane_recenice)
Konkordancija.concordance("thi")
Konkordancija.concordance("movi")
Konkordancija.concordance("film")
Konkordancija.concordance("n't")
Konkordancija.concordance("like")
Konkordancija.concordance("one")
Konkordancija.concordance("hi")
Konkordancija.concordance("watch")
Konkordancija.concordance("bad")
Konkordancija.concordance("even")
```

```python
#Prikazivanje kolokacije za 3 najcesce rijeci u teksu
biagram_collocation = BigramCollocationFinder.from_words(filtrirane_recenice) 
biagram_collocation.nbest(BigramAssocMeasures.likelihood_ratio, 50) 
    
```

```python

review_text = Text(filtrirane_recenice)

def lexical_diversity(txt):
    return len(txt) / len(set(txt))

lexical_diversity(review_text)

print(lexical_diversity(filtrirane_recenice))

print(lexical_diversity(raw))



```

```python
#Primjena korpusa teksta npr Brown, broj ponavaljanja odabranih rijeci u odabranoj kategoriji


browntekst = movie_reviews.words()


#Micanje stop word-ova i punktacije
stopWords = set(stopwords.words('english'))
punctuation = list(string.punctuation)
dodatni_znakovi = ['[', '}', '{',']','br','—','…','–','``',"...","''"]

punctuation.extend(dodatni_znakovi)
browntekstfiltriran = []
for x in browntekst:
    if x.lower() not in stopWords and x.lower() not in punctuation:
        browntekstfiltriran.append(x)
        
frekvencija = nltk.FreqDist(x.lower() for x in browntekstfiltriran)

#Najčešće riječi
print(frekvencija.most_common(30))
#usporedba s brownom
print(lexical_diversity(browntekst))
biagramKolokacija = BigramCollocationFinder.from_words(browntekstfiltriran) 
biagramKolokacija.nbest(BigramAssocMeasures.likelihood_ratio, 50) 
```
